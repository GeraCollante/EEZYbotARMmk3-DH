```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go

%matplotlib inline
from numpy import *
d2r = np.deg2rad
r2d = np.degrees
```

```python
from sympy.physics.mechanics import dynamicsymbols
```

### Parámetros DH

```python
theta, alpha, a, d = dynamicsymbols('theta alpha a d')
theta, alpha, a, d 
```

### Parámetros del robot

```python
theta1, theta2, theta3 = dynamicsymbols('theta1 theta2 theta3')
theta1, theta2, theta3
```

```python
rot = sp.Matrix([[sp.cos(theta), -sp.sin(theta)*sp.cos(alpha), sp.sin(theta)*sp.sin(alpha)],
                 [sp.sin(theta), sp.cos(theta)*sp.cos(alpha), -sp.cos(theta)*sp.sin(alpha)],
                 [0, sp.sin(alpha), sp.cos(alpha)]])

trans = sp.Matrix([a*sp.cos(theta),a*sp.sin(theta),d])

last_row = sp.Matrix([[0, 0, 0, 1]])

m = sp.Matrix.vstack(sp.Matrix.hstack(rot, trans), last_row)
m
```

```python
l1 = 0.2
l2 = 0.5
L1 = 0.7
```

```python
m01 = m.subs({theta:theta1, d:0.2, a:0, alpha:pi/2})
m12 = m.subs({theta:pi/2, d:0, a:0.7, alpha:0})
m23 = m.subs({theta:theta3, d:0, a:1.2, alpha:-pi/2})
m34 = m.subs({theta:pi/2, d:0, a:0, alpha:pi/2})

m03 = (m01*m12*m23*m34)
#m03 = (m01)
mtcp = sp.simplify(m03)
px = mtcp[0,3]
py = mtcp[1,3]
pz = mtcp[2,3]
fx = sp.lambdify((theta1, theta2, theta3), px, 'numpy')
fy = sp.lambdify((theta1, theta2, theta3), py, 'numpy')
fz = sp.lambdify((theta1, theta2, theta3), pz, 'numpy')
```

```python
t1 = r2d(0)
t2 = r2d(45)
mtcp = sp.simplify(m03)
px = mtcp[:3,0]
py = mtcp[:3,1]
pz = mtcp[:3,2]
u = px.subs([(theta1, t1), (theta2,t2)])
v = py.subs([(theta1, t1), (theta2,t2)])
w = pz.subs([(theta1, t1), (theta2,t2)])
p12 = tcp.subs([(theta1, 0), (theta2,t2)])
```

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
from matplotlib.pyplot import figure

import matplotlib.pyplot as plt
import numpy as np

def set_axes_equal(ax):
    '''Make axes of 3D plot have equal scale so that spheres appear as spheres,
    cubes as cubes, etc..  This is one possible solution to Matplotlib's
    ax.set_aspect('equal') and ax.axis('equal') not working for 3D.

    Input
      ax: a matplotlib axis, e.g., as output from plt.gca().
    '''

    x_limits = ax.get_xlim3d()
    y_limits = ax.get_ylim3d()
    z_limits = ax.get_zlim3d()

    x_range = abs(x_limits[1] - x_limits[0])
    x_middle = np.mean(x_limits)
    y_range = abs(y_limits[1] - y_limits[0])
    y_middle = np.mean(y_limits)
    z_range = abs(z_limits[1] - z_limits[0])
    z_middle = np.mean(z_limits)

    # The plot bounding box is a sphere in the sense of the infinity
    # norm, hence I call half the max range the plot radius.
    plot_radius = 0.5*max([x_range, y_range, z_range])

    ax.set_xlim3d([x_middle - plot_radius, x_middle + plot_radius])
    ax.set_ylim3d([y_middle - plot_radius, y_middle + plot_radius])
    ax.set_zlim3d([z_middle - plot_radius, z_middle + plot_radius])

fig = plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
ax = fig.gca(projection='3d')
'''
# Make the grid
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Make the direction data for the arrows
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

x, y, z = np.array([0,0,0])
#u, v, w = np.array([4,5,8])
'''
ax.quiver(0, 0, 0, 1, 0, 0, color='r', length=0.1, normalize=True, label='$x_0$')
ax.quiver(0, 0, 0, 0, 1, 0, color='g', length=0.1, normalize=True, label='$y_0$')
ax.quiver(0, 0, 0, 0, 0, 1, color='b', length=0.1, normalize=True, label='$z_0$')

ax.quiver(p01[0], p01[1], p01[2], u[0], u[1], u[2], color='r', length=0.1, normalize=True)
ax.quiver(p01[0], p01[1], p01[2], v[0], v[1], v[2], color='g', length=0.1, normalize=True)
ax.quiver(p01[0], p01[1], p01[2], w[0], w[1], w[2], color='b', length=0.1, normalize=True)
ax.scatter(p01[0], p01[1], p01[2])

ax.legend()

set_axes_equal(ax)

plt.tight_layout()
plt.show()
```
